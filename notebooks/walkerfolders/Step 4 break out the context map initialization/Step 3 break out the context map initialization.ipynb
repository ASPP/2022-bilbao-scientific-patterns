```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

from walker import Walker, hills_context_map_builder, flat_context_map_builder, labyrinth_context_map_builder, plot_trajectory
```

# Simulate a trajectory in a hills context

```python
# Create a Walker instance
walker = Walker.from_context_map_builder(
    sigma_i=3,
    sigma_j=4,
    size=200,
    context_map_builder=hills_context_map_builder,
)

# Sample a next step 1000 times
i, j = 100, 50
trajectory = []
for _ in range(1000):
    i, j = walker.sample_next_step(i, j)
    trajectory.append((i, j))
```

```python
plot_trajectory(trajectory, walker.context_map)
```
